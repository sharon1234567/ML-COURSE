```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'http://www.bjsubway.com/e/action/ListInfo/?classid=39&ph=1'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text,"html5lib")
```

```python
print(res.text)
```

```python
def valid_info(string):
    return ''.join(re.findall(r'.*',string))
p = re.compile(r'<th>\s*[\u4e00-\u9fa5]+</th>|<th colspan="2">\s*[\u4e00-\u9fa5]+</th>|<th style="background:#DEAF90;">\s*[\u4e00-\u9fa5]+</th>|<td .*>.*首末车时刻表</td>')
stations=[]
#info_line = soup.find_all('td')
info = soup.find_all('tr')
for i in range(len(info)):
    print(i,str(info[i]))
    station = p.findall(valid_info(str(info[i])))
    if station:
        #print(station)
        stations.append(station)
```

```python
stations
```

```python
station_parsed = []
p2 = re.compile('\w*[0-9]*[\u4e00-\u9fa5]+')
for i in range(len(stations)):
    if '时间' not in stations[i][0] and '方向' not in stations[i][0]:
        station = stations[i][0]
        station_parsed.append(''.join(p2.findall(station)))
```

```python
station_parsed
```

```python
import pandas as pd
df_station = pd.DataFrame(columns=['Name','Stations'])
```

```python
df_station
```

```python
station_list = []
k = 0
while k<len(station_parsed):
    if '首末车时刻表' in station_parsed[k]:
        k+=1
        if station_list:
            df_station = df_station.append([{'Name':station_parsed[k-len(station_list)-2],'Stations':station_list}])
        station_list = []
        
    else:
        station_list.append(station_parsed[k])
        k+=1
df_station = df_station.append([{'Name':station_parsed[k-len(station_list)-1],'Stations':station_list}])
```

```python
df_station
```

```python
import networkx as nx
from collections import defaultdict
import matplotlib as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

station_set =  set(station_parsed)

stations_connection = defaultdict(list)
for s1 in station_set:
    for s2 in station_set:
        if s1==s2: continue
        for i in range(len(df_station)):
            if s1 not in df_station.iat[i,1] or s2 not in df_station.iat[i,1]: continue
            elif abs(df_station.iat[i,1].index(s1)-df_station.iat[i,1].index(s2))==1 and s2 not in stations_connection[s1]:
                stations_connection[s1].append(s2)
            break
                   
```

```python
stations_connection_graph  = nx.Graph(stations_connection)
```

```python
import matplotlib.pyplot as plt
plt.figure(1,figsize=(12,12))
nx.draw(stations_connection_graph, with_labels=True, node_size=10)
```

```python
def search(graph, start, goal, strategy):
    pathes = [[start]]
    seen = set()
    
    while pathes:
        path = pathes.pop(0)
        #print('path:',path)
        frontier = path[-1]
        #print('frontier:',frontier)
        
        if frontier in seen: continue
        
        successors = graph[frontier]
        #print('successors:',successors)
        
        for city in successors:
            if city in path: continue
            
            new_path = path+[city]
            if goal in new_path:
                #print('Done')
                return new_path
            
            pathes.append(new_path)
        #print('pathes:', pathes)
            
        seen.add(frontier)
        pathes = strategy(pathes)
        #print('pathes_sorted:', pathes)
        
def search_byway(graph, start, byway, goal, strategy):
    all_line = [start] + byway +[goal]
    path = []
    for i in range(len(all_line)-1):
        sub_path =  search(graph, all_line[i], all_line[i+1], strategy)
        path += sub_path[1:]
    return path
            
def get_total_station(path):
    return len(path)

def get_change_times(path):
    times = 1
    for i in range(len(path)-2):
        if in_the_same_line(path[i],path[i+2])==False:
            times += 1
    return times
            
def in_the_same_line(s1,s2):
    for i in range(len(df_station)):
        if (s1 in df_station.iat[i,1])and (s2 in df_station.iat[i,1]):
            return True
    return False

def sort_path(cmp_func, beam=-1):
    def _sorted(pathes):
        return sorted(pathes, key=cmp_func)
    return _sorted
    
```

```python
search(stations_connection_graph, '大井', '莲花桥', strategy = sort_path(get_total_station))
```
